````python
import numpy
r_ab = ap_pos[:,numpy.newaxis] - all_pos[numpy.newaxis,:]
print(r_ab.shape)
````

````python
d_ab = numpy.sqrt(numpy.einsum('ijk,ijk->ij', r_ab, r_ab))
print(d_ab.shape)
````

````python
apolar_clash = numpy.logical_and(d_ab < 2.9, d_ab > 0)
print(apolar_clash.shape)
````
